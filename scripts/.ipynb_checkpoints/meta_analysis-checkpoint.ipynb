## 2.1 Quality control

Here, we perfrom both cell and gene QC. In order to assess the quality of a cell we observe the number of expressed genes and the fraction of counts that are mitochondrial, whereas gene quality is assessed by cell count i.e detection in small or large number of cells.  `A high fraction of mitochondrial reads being picked up can indicate cell stress, as there is a low proportion of nuclear mRNA in the cell`. The high proportions of mitochondrial genes, indicative of poor-quality cells [(Lun, McCarthy & Marioni, 2017)](https://master.bioconductor.org/packages/release/workflows/html/simpleSingleCell.html#examining-gene-level-metrics), could also likley occurs as a result of loss of cytoplasmic RNA from perforated cells. Since mitochondria are larger than individual transcript molecules and less likely to escape through tears in the cell membrane. It should be noted that high mitochondrial RNA fractions can also be biological signals indicating elevated respiration. 

Since we are working with filtered/proprocessed data, we set a very low threshold for these initial filtering steps, to avoid filtering out informative genes or cells.

To assemble some of the information/metrics for quality control, we employ the `pp.calculate_qc_metrics`, in `scanpy`.

Obtain QC metrics for `Leng ETX`

```python
adata_leng_etx.var['mt'] = adata_leng_etx.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(adata_leng_etx, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

Basic Filtering

```python
def basic_filter(adata: dict,
                min_genes: int = 100,
                min_cells: int = 3):
    for key in adata.keys():
        sc.pp.filter_cells(adata[key], min_genes=min_genes)
        sc.pp.filter_genes(adata[key], min_cells=min_cells)

    return adata
```

```python
adata_query = basic_filter(adata_query)
```

A violin plot of some of the computed quality measures:

* the number of genes expressed in the count matrix
* the total counts per cell
* the percentage of counts in mitochondrial genes

```python
for key in adata_query.keys():
    print(f'Plotting QC metrics for for {key.upper()} data')
    sc.pl.violin(adata_query[key], ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
     jitter=0.9, multi_panel=True)
    print('\n')
```

The plots show that the `grubman_etx` data has the poorest quality in comparison. For all data sources, the counts per cell are densely distributed with a significant number of cells low`pct_counts_mt` where the fraction of mitochondrial reads are below 10-15%. We subsequently set `10%` ad the filtering threshold for all data.

However, we perform filtering of loq quality cells using the `pct_counts_mt` covariate in isolation can lead to misinterpretation of the underlying cellular information. To avoid this, we jointly consider the covariates using the following plots.

```python
# Data quality summary plots
for key in adata_query.keys():
    print(f'Plotting QC metrics for for {key.upper()} data')
    p1 = sc.pl.scatter(adata_query[key], 'total_counts', 'n_genes_by_counts', color='pct_counts_mt')
    #p2 = sc.pl.scatter(adata_query[key][adata_query[key].obs['n_counts']<10000], 'n_counts', 'n_genes', color='mt_frac')
    print('\n')
```

From the plots above, we can see that there are only a handful of cells with unexpected summary statistics. For example, in the `LENG_ETX` data, there's a low density of cells with high total counts and number of detected genes. These cells also have very low fraction of mitochondrial genes detected. Therefore, we filter out cells with unexpected (high total counts and number of detected genes) from each of the datasets.

`..cells with lower counts and genes tend to have a higher fraction of mitochondrial counts. These cells are likely under stress or are dying. When apoptotic cells are sequenced, there is less mRNA to be captured in the nucleus, and therefore fewer counts overall, and thus a higher fraction of counts fall upon mitochondrial RNA. If cells with high mitochondrial activity were found at higher counts/genes per cell, this would indicate biologically relevant mitochondrial activity.` [Malte D Luecken1 & Fabian J Theis](https://www.embopress.org/doi/full/10.15252/msb.20188746)

Although the `GRUBMAN_ETX` data shows relatively higher mitochondrial fraction, we refrain from further filtering since `a recent systematic survey of scRNA-seq data suggested that a mitochondrial proportion threshold of 10% is appropriate to distinguish between healthy and low-quality cells in most human tissues, while in mouse tissues, the recommended threshold is 5%` [Osorio D & Cai JJ](https://academic.oup.com/bioinformatics/article/37/7/963/5896986?login=false)


To assert the thresholding descisions, we visualize the library size accross all datasets. 

```python
#Thresholding decision: counts
for key in adata_query.keys():
    print(f'Plotting QC metrics for for {key.upper()} data')
    p3 = sb.distplot(adata_query[key].obs['total_counts'], kde=False)
    plt.show()
    print('\n')
```

The above plots show that a uniform distribution of feature (gene) expression in most cells. However, we note that total counts in the `GRUBMAN_ETX` data has two peaks (bimodal) , indicative of two expression profiles in the population. We take this into account to futher assess the cells with total counts less than 2000. 

```python
#Thresholding decision: counts
for key in adata_query.keys():
    print(f'Plotting QC metrics for for {key.upper()} data')
    p3 = sb.distplot(adata_query[key].obs['total_counts'][adata_query[key].obs['total_counts']<2000], kde=False)
    plt.show()
    print('\n')
```

From the above hsitogram distributions and cummulative plots of estimated cell QC metrics, it is apparent that certain cells have unusual distributions. Cells with very few expressed features or small library size indicate low RNA-capture efficiency and are hence considered poor quality. On the other hand, cells with abnormally a large number of expressed features suggest doublets or multiplets.

Therefore we estimate the doublets/multiplets present in each data before further filtering is performed. To do this, we use the `Scrublet` and `DoubletDetection` doublet prediction python packages which have been shown to demonstrate strong performance compared with the other doublet detection methods and particularly achieved top performance in removing spurious cell clusters, homotypic doublets, and inferred trajectories that most resembled the trajectory inferred from the clean data.

Although functionality for assessing `DoubletDetection` predictions is provided below, we stick with results from `Srublet` since it is approximately `4 times faster` than `DoubletDetection`.

##### Scrublet

```python
for key in adata_query.keys():
    print(f'Estimating doublets on {key} data using Scrublet')
    sc.external.pp.scrublet(adata_query[key], sim_doublet_ratio=2.0, 
                        expected_doublet_rate=0.05, stdev_doublet_rate=0.02,
                        knn_dist_metric='euclidean', n_prin_comps=30, get_doublet_neighbor_parents=True,
                        n_neighbors=None, threshold=None, verbose=False, random_state=0)

    sc.external.pl.scrublet_score_distribution(adata_query[key], scale_hist_obs='log',
                                               scale_hist_sim='linear', figsize=(8, 3),
                                               return_fig=False, show=True, save=None)
    print("\n")
```

##### `Leng ETX`

`Pre-filtering Metrics`

```python
# Data quality summary plots

fig, axs = plt.subplots(1, 2, figsize=(20, 8))
key = 'leng_etx'

x = adata_query[key].obs['total_counts']
y = adata_query[key].obs['n_genes_by_counts']
doublet_color = adata_query[key].obs['doublet_score']
mt_count_color = adata_query[key].obs['pct_counts_mt']

doublet_axes = axs[0].scatter(x, y, c=doublet_color,)
axs[0].set_xlabel('total_counts')
axs[0].set_ylabel('n_genes_by_counts')
axs[0].set_title(f'doublet_score for {key}')
plt.colorbar(doublet_axes)

mt_axes = axs[1].scatter(x, y, c=mt_count_color, )
axs[1].set_xlabel('total_counts')
axs[1].set_ylabel('n_genes_by_counts')
axs[1].set_title(f'pct_counts_mt for {key}')
plt.colorbar(mt_axes)
```

From the plots above, the plots above, we see that although there are cells with unusually high distributions, they are likley high quality or informative cells as they do not exhibit high doublet scores or mitochondrial fraction. In contrast, we observe a unique cluster of cells with uusual distribution of cells and relatively high mitochondrial fraction.

Taking these observations into consideration, we filter only cells with `pct_counts_mt > 10%` since previous review of acceptable threshold suggests that that a mitochondrial proportion threshold of 10% is sufficient to distinguish between healthy and low-quality cells in most human tissues[Osorio D & Cai JJ](https://academic.oup.com/bioinformatics/article/37/7/963/5896986?login=false)

```python
# Filter cells according to identified QC thresholds:

key = 'leng_etx'
print('Total number of {} cells: {:d}'.format(key, adata_query[key].n_obs))

adata_query[key] = adata_query[key][adata_query[key].obs['pct_counts_mt'] < 10]
print('Number of {} cells after MT filter: {:d}'.format(key, adata_query[key].n_obs))
```

`Post-filtering Metrics`

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 8))
key = 'leng_etx'

x = adata_query[key].obs['total_counts']
y = adata_query[key].obs['n_genes_by_counts']
doublet_color = adata_query[key].obs['doublet_score']
mt_count_color = adata_query[key].obs['pct_counts_mt']

doublet_axes = axs[0].scatter(x, y, c=doublet_color,)
axs[0].set_xlabel('total_counts')
axs[0].set_ylabel('n_genes_by_counts')
axs[0].set_title(f'doublet_score for {key}')
plt.colorbar(doublet_axes)

mt_axes = axs[1].scatter(x, y, c=mt_count_color, )
axs[1].set_xlabel('total_counts')
axs[1].set_ylabel('n_genes_by_counts')
axs[1].set_title(f'pct_counts_mt for {key}')
plt.colorbar(mt_axes)
```
